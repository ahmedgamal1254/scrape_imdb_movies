```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.imdb.com/chart/toptv/'
source=requests.get(url)
```

```python
print(source.raise_for_status())
```

```python
html=source.content
```

```python
soup=BeautifulSoup(html,"html.parser")
soup
```

```python
title_movie=soup.findAll("td",{"class":"titleColumn"})
len(title_movie)
```

```python
title_movie[0].find("a").text
```

```python
date=title_movie[0].find("span").text
date
```

```python
rating=soup.findAll("td",{"class":"ratingColumn imdbRating"})
len(rating)
```

```python
rating[0].find("strong").text
```

```python
parent_of_movies=soup.find("tbody",{"class":"lister-list"})
```

```python
movies=parent_of_movies.findAll("tr")
len(movies)
```

```python
movies[0]
```

```python
movies[0].find("td",{"class":"titleColumn"}).a.text
```

```python
history=movies[0].find("td",{"class":"titleColumn"}).span.text
history
```

```python
movies[0].find("td",{"class":"ratingColumn"}).strong.text
```

```python
import re
history=re.search("[0-9]{4}",history)[0]
history
```

```python
rank=movies[0].find("td",{"class":"titleColumn"}).get_text(strip=True).split(".")[0]
rank
```

```python
import csv
wr=open("imdb_2.csv","w")
writer=csv.writer(wr)
writer.writerow(["Rank","Title","Year","Rate"])
wr.close()
```

```python
for movie in movies:
    rank=movie.find("td",{"class":"titleColumn"}).get_text(strip=True).split(".")[0]
    title=movie.find("td",{"class":"titleColumn"}).a.text
    history=movie.find("td",{"class":"titleColumn"}).span.text
    history=re.search("[0-9]{4}",history)[0]
    rate=movie.find("td",{"class":"ratingColumn"}).strong.text
    
    wr=open("imdb_2.csv","a",newline='')
    writer=csv.writer(wr)
    writer.writerow([rank,title,history,rate])
    wr.close()
```
